**option("cloudFiles.schemaLocation", "/Volumes/.../schema")**

-->this is for store inferschema and metadata

**.option("cloudFiles.maxFilesPerTrigger", 1)**

-->this is for processing each file at one point 

**.option("cloudFiles.schemaEvolutionMode", "rescue")**

-->this for handle schema change

  Other modes:

    failOnNewColumns 

    addNewColumns 

```python
    df= spark.readStream.format("cloudFiles") \
        .option("cloudFiles.format", "csv") \
        .option("cloudFiles.schemaLocation", "/Volumes/pyspark_catalog_scinario_questions/auto_load/destination/schema") \
        .option("cloudFiles.maxFilesPerTrigger", 1) \
        .option("cloudFiles.schemaEvolutionMode", "rescue")\
        .load("/Volumes/pyspark_catalog_scinario_questions/auto_load/source/")
```

**.option("checkpointLocation", ".../checkpoint")**

-->it stores meta data which file processed

**.trigger(once=True)**

--> this means process file available this query execute one time

```python
df.writeStream.format("delta")\
    .option("checkpointLocation", "/Volumes/pyspark_catalog_scinario_questions/auto_load/destination/checkpoint")\
    .trigger(once=True)\
    .start("/Volumes/pyspark_catalog_scinario_questions/auto_load/destination/data/")
```
